# Dataset Detail

```python
import numpy as np
import pandas as pd
import os 
import matplotlib.pyplot as plt
import shutil
```

```python
file_path = 'E:\\AgeDB-1'
l  = []
for dirname, _, filenames in os.walk(file_path):
    for filename in filenames:
        filename = filename.replace('.jpg', '')
        filename = filename.split('_')
        l.append(filename)
```

```python
filename
```

```python
data = pd.DataFrame(l)
data.columns = ["no","name","age","gender"]
data["name"] = data["name"].str.capitalize()
```

```python
data
```

```python
def func(df):
    return ','.join(df.values)

data = data.groupby(by='name').agg(func).reset_index()
data.drop(["no","gender"],axis=1,inplace=True)
```

```python
data
```

```python
name_list = []
gender_list = []
for i in range(len(l)):
    name = str.capitalize(l[i][1]) # captilize
    gender = l[i][3]
    if name not in name_list:
        name_list.append(name)
        gender_list.append(gender)
        
df = pd.DataFrame(name_list)
df.columns = ["name"]
df["gender"] = gender_list
```

```python
df_merge = pd.merge(df, data, on = 'name')
df_merge["total_images"] = ""
df_merge["0-10"] = ""
df_merge["11-20"] = ""
df_merge["21-30"] = ""
df_merge["31-40"] = ""
df_merge["41-50"] = ""
df_merge["51-60"] = ""
df_merge["61-70"] = ""
df_merge["71+"] = ""
```

```python
for j in range(df_merge.shape[0]):
    n = df_merge.iat[j,2]
    n1 = n.split(',')
    r1 = 0
    r2 = 0
    r3 = 0
    r4 = 0
    r5 = 0
    r6 = 0
    r7 = 0
    r8 = 0
    total = 0 
    for i in n1:
        i = int(i)
        if i <= 10:
            r1 += 1
        if i > 10 and i<= 20:
            r2 += 1
        if i > 20 and i<= 30:
            r3 += 1
        if i > 30 and i<= 40:
            r4 += 1
        if i > 40 and i<= 50:
            r5 += 1
        if i > 50 and i<= 60:
            r6 += 1
        if i > 60 and i<= 70:
            r7 += 1
        if i > 70:
            r8 += 1
        total += 1
    df_merge.iat[j,3] = total
    df_merge.iat[j,4] = r1
    df_merge.iat[j,5] = r2
    df_merge.iat[j,6] = r3
    df_merge.iat[j,7] = r4
    df_merge.iat[j,8] = r5
    df_merge.iat[j,9] = r6
    df_merge.iat[j,10] = r7
    df_merge.iat[j,11] = r8
pd.set_option('display.max_rows', None) 
df_merge
```

```python
df_merge = df_merge.sort_values(by=['name'],ignore_index=True)
df_merge
```

# Change the file name

```python
import os
import shutil
# construct every folder for every  person
path = "E:\\AgeDB-1-copy"
 
files_list = os.listdir(path)
 
for file in files_list:
    name1 = file.split('_', 3)[1]
    name1 = name1.replace(" ","")

    if not os.path.exists(os.path.join(path, name1)):
        os.makedirs(os.path.join(path, name1))
 
    shutil.move(os.path.join(path, file), os.path.join(path, name1))
```

```python
# delete 7 folders [21,72,85,147,194,381,405]
for file in [[21,72,85,147,194,381,405]:
     folder_path = os.path.join(path,file)
     os.remove(folder_path) 
```

```python
# if we would like to  rename the picture name, we find, they have same age so same age
file = os.listdir(path)
num = 0

for i in range(len(file)):
    n1 = path + '\\' + file[i]
    n2 = path + '\\' + str(num) 
    os.rename(n1, n2)
    num += 1
```

```python
for i in file:
    path = 'E:\\AgeDB-copy\\' + str(i)
    files_list = os.listdir(path)
 
    for file in files_list:
        filename = path + '\\' + file
#         print(filename)
        new_name = file.split('_', 3)[2]
        new_name = new_name.replace(" ","")
        n = path + '\\' + new_name + '.jpg'
        if os.path.exists(n) == True:
            continue
        os.rename(filename, n)        
```
