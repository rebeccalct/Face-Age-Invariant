```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os 
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

## Load Gender Information

```python
file_path = 'E:\\AgeDB-1'
l  = []
for dirname, _, filenames in os.walk(file_path):
    for filename in filenames:
        filename = filename.replace('.jpg', '')
        filename = filename.split('_')
        l.append(filename)
```

```python
data = pd.DataFrame(l)
data.columns = ["no","name","age","g"]
data["name"] = data["name"].str.capitalize()
```

```python
def func(df):
    return ','.join(df.values)

data = data.groupby(by='name').agg(func).reset_index()
data['gender'] = data['g'].map(lambda x:x.split(',')[0])
data.drop(["no","g","age"],axis=1,inplace=True)
data.index.name = 'Person'
data
```

```python
data["gender"].describe()
```

## Compare Average in Same Age Range

```python
filename_path2 = 'C:\\Users\\rebeccalai\\OneDrive\\desktop\\FACE\\result\\insight1.csv'
insight1 = pd.read_csv(filename_path2)
```

```python
df = pd.merge(insight1, data, on='Person')
df
```

```python
df_f = df.loc[df["gender"] == 'f']
df_f.loc['female average'] = df_f.loc[:,"<=10":"70+"].apply(lambda x:x.mean())
df_m = df.loc[df["gender"] == 'm']
df_m.loc['male average'] = df_m.loc[:,"<=10":"70+"].apply(lambda x:x.mean())
```

```python
df_m
```

```python
dataframe1 = pd.DataFrame(columns=df.columns)
dataframe1.loc['male average'] = df_m.loc['male average']
dataframe1.loc['female average'] = df_f.loc['female average']
dataframe1 = dataframe1.drop(['Unnamed: 0','Person','name','gender'], axis = 1)
dataframe1
```

```python
x = dataframe1.columns
y1 = dataframe1.loc['male average']
y2 = dataframe1.loc['female average']
plt.plot(x,y2,'.b-', label="female")
plt.plot(x,y1,'.r-', label="male")

plt.legend(loc=2, bbox_to_anchor=(1.05,1.0),borderaxespad = 0.)
plt.xlabel("Age-range")
plt.ylabel("Mean Similarity")
plt.show()
```
